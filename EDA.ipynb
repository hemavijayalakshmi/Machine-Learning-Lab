```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define custom colors for each species
custom_palette = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}

# Creating the count plot with custom colors and avoiding the warning
sns.countplot(x='species', data=df, hue='species', palette=custom_palette, legend=False)

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot(x='sepal_length', y='sepal_width',
                hue='species', data=df, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot(x='petal_length', y='petal_width',
                hue='species', data=df, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```
